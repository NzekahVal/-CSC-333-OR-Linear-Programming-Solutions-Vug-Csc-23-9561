```python
# importing necessary libraries for proper code execution
import numpy as np
import matplotlib.pyplot as plt
"""
MAXIMIZE PROFIT
            Z=6x+7y
SUBJECT TO

            3x+4y≤18
            2x+y≤10
            x≥0,y≥0 (non-negativity constraints)
"""
# Define the constraints for the transport company problem
def fuel_constraint(x):
    return (18 - 3 * x) / 4  # From 3x + 4y = 18

def driver_time_constraint(x):
    return (10 - 2 * x)  # From 2x + y = 10

# Generate x values for the plot
x = np.linspace(0, 6, 400)

# Calculate y values for each constraint
y_fuel = fuel_constraint(x)
y_driver_time = driver_time_constraint(x)

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot the constraint lines
plt.plot(x, y_fuel, label=r'$3x + 4y \leq 18$', color='blue')
plt.plot(x, y_driver_time, label=r'$2x + y \leq 10$', color='green')

# Highlight the feasible region
y_fill = np.minimum(y_fuel, y_driver_time)
plt.fill_between(x, y_fill, 0, where=(y_fill >= 0), color='gray', alpha=0.3, label='Feasible Region')

# Set axis limits
plt.xlim(0, 6)
plt.ylim(0, 6)

# Label the plot
plt.title('Graphical Solution to Minimize Transport Cost')
plt.xlabel('Trips by Vehicle X (x)')
plt.ylabel('Trips by Vehicle Y (y)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
