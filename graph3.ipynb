```python
# importing necessary libraries for proper code execution
import numpy as np
import matplotlib.pyplot as plt

"""
MAXIMIZE
            Z=5x+4y
SUBJECT TO

            2x+y≤20
            3x+2y≤30
            x+2y≤18
            x≥0,y≥0 (non-negativity constraints)
"""
# Define the constraint equations
def labor_constraint(x):
    return (20 - 2*x)  # From 2x + y = 20

def material_constraint(x):
    return (30 - 3*x) / 2  # From 3x + 2y = 30

def machine_time_constraint(x):
    return (18 - x) / 2  # From x + 2y = 18

# Generate x values for the plot
x = np.linspace(0, 15, 400)

# Calculate y values for each constraint
y1 = labor_constraint(x)
y2 = material_constraint(x)
y3 = machine_time_constraint(x)

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot the constraint lines
plt.plot(x, y1, label=r'$2x + y \leq 20$', color='blue')
plt.plot(x, y2, label=r'$3x + 2y \leq 30$', color='green')
plt.plot(x, y3, label=r'$x + 2y \leq 18$', color='orange')

# Highlight the feasible region
y_fill = np.minimum(np.minimum(y1, y2), y3)
plt.fill_between(x, y_fill, 0, where=(y_fill >= 0), color='gray', alpha=0.3, label='Feasible Region')

# Set axis limits
plt.xlim(0, 12)
plt.ylim(0, 12)

# Label the plot
plt.title('Graphical Solution to Maximize Profit')
plt.xlabel('Units of Product A (x)')
plt.ylabel('Units of Product B (y)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
