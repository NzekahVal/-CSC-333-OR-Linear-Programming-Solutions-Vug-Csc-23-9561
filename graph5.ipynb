```python
# importing necessary libraries for proper code execution
import numpy as np
import matplotlib.pyplot as plt
"""
MAXIMIZE
            Z=8x+7y
SUBJECT TO

            3x+4y≤12
            2x+y≤6
            x≥0,y≥0 (non-negativity constraints)
"""
# Define the constraints
def labor_constraint(x):
    return (12 - 3 * x) / 4  # From 3x + 4y = 12

def capital_constraint(x):
    return (6 - 2 * x)  # From 2x + y = 6

# Generate x values for the plot
x = np.linspace(0, 6, 400)

# Calculate y values for each constraint
y_labor = labor_constraint(x)
y_capital = capital_constraint(x)

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot the constraint lines
plt.plot(x, y_labor, label=r'$3x + 4y \leq 12$', color='blue')
plt.plot(x, y_capital, label=r'$2x + y \leq 6$', color='green')

# Highlight the feasible region
y_fill = np.minimum(y_labor, y_capital)
plt.fill_between(x, y_fill, 0, where=(y_fill >= 0), color='gray', alpha=0.3, label='Feasible Region')

# Set axis limits
plt.xlim(0, 5)
plt.ylim(0, 5)

# Label the plot
plt.title('Graphical Solution to Maximize Profit')
plt.xlabel('Units of Project P1 (x)')
plt.ylabel('Units of Project P2 (y)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
