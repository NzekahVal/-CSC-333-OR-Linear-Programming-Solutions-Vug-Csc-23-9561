```python
# importing necessary libraries for proper code execution
import numpy as np
import matplotlib.pyplot as plt

"""
MINIMIZE
            Z=2x+5y
SUBJECT TO

            x+2y≤6
            2x+y≤5
            x≥0,y≥0 (non-negativity constraints)
"""
# Define the constraint equations
def constraint1(x):
    return (6 - x) / 2  # From x + 2y = 6

def constraint2(x):
    return (5 - 2*x)  # From 2x + y = 5

# Generate x values for the plot
x = np.linspace(0, 5, 400)

# Calculate y values for each constraint
y1 = constraint1(x)
y2 = constraint2(x)

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot the constraint lines
plt.plot(x, y1, label=r'$x + 2y \leq 6$', color='blue')
plt.plot(x, y2, label=r'$2x + y \leq 5$', color='green')

# Highlight the feasible region
y_fill = np.minimum(y1, y2)
plt.fill_between(x, y_fill, 0, where=(y_fill >= 0), color='gray', alpha=0.3, label='Feasible Region')

# Set axis limits
plt.xlim(0, 4)
plt.ylim(0, 3)

# Label the plot
plt.title('Graphical Solution to Minimize Production Cost')
plt.xlabel('Units of Product X (x)')
plt.ylabel('Units of Product Y (y)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
