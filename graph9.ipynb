```python
# importing necessary libraries for proper code execution
import numpy as np
import matplotlib.pyplot as plt

"""
MAXIMIZE PROFIT
            Z=500000x+400000y
SUBJECT TO

            4000x+3000y≤5000
            2000x+2500y≤4500
            1000x+1500y≤3000
            x≥0,y≥0 (non-negativity constraints)
"""
# Define the constraints for the advertising campaigns
def tv_budget_constraint(x):
    return (5000 - 4000 * x) / 3000  # From 4000x + 3000y <= 5000

def print_media_budget_constraint(x):
    return (4500 - 2000 * x) / 2500  # From 2000x + 2500y <= 4500

def social_media_budget_constraint(x):
    return (3000 - 1000 * x) / 1500  # From 1000x + 1500y <= 3000

# Generate x values for the plot
x = np.linspace(0, 2, 400)

# Calculate y values for each constraint
y_tv = tv_budget_constraint(x)
y_print_media = print_media_budget_constraint(x)
y_social_media = social_media_budget_constraint(x)

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot the constraint lines
plt.plot(x, y_tv, label=r'$4000x + 3000y \leq 5000$', color='blue')
plt.plot(x, y_print_media, label=r'$2000x + 2500y \leq 4500$', color='green')
plt.plot(x, y_social_media, label=r'$1000x + 1500y \leq 3000$', color='red')

# Highlight the feasible region
y_fill = np.minimum(np.minimum(y_tv, y_print_media), y_social_media)
plt.fill_between(x, y_fill, 0, where=(y_fill >= 0), color='gray', alpha=0.3, label='Feasible Region')

# Set axis limits
plt.xlim(0, 2)
plt.ylim(0, 2)

# Label the plot
plt.title('Graphical Solution to Maximize Reach of Advertising Campaigns')
plt.xlabel('Units of Campaign A (x)')
plt.ylabel('Units of Campaign B (y)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
