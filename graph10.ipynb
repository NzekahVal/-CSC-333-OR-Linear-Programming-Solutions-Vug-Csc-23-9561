```python
# importing necessary libraries for proper code execution
import numpy as np
import matplotlib.pyplot as plt

"""
MAXIMIZE PROFIT
            Z=6x+5y
SUBJECT TO

            2x+4y≤30
            3x+2y≤24
            x+2y≤20
            x≥0,y≥0 (non-negativity constraints)
"""
# Define the constraints for the cafeteria meal problem
def meat_constraint(x):
    return (30 - 2 * x) / 4  # From 2x + 4y <= 30

def vegetable_constraint(x):
    return (24 - 3 * x) / 2  # From 3x + 2y <= 24

def rice_constraint(x):
    return (20 - x) / 2  # From x + 2y <= 20

# Generate x values for the plot
x = np.linspace(0, 12, 400)

# Calculate y values for each constraint
y_meat = meat_constraint(x)
y_vegetable = vegetable_constraint(x)
y_rice = rice_constraint(x)

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot the constraint lines
plt.plot(x, y_meat, label=r'$2x + 4y \leq 30$', color='blue')
plt.plot(x, y_vegetable, label=r'$3x + 2y \leq 24$', color='green')
plt.plot(x, y_rice, label=r'$x + 2y \leq 20$', color='red')

# Highlight the feasible region
y_fill = np.minimum(np.minimum(y_meat, y_vegetable), y_rice)
plt.fill_between(x, y_fill, 0, where=(y_fill >= 0), color='gray', alpha=0.3, label='Feasible Region')

# Set axis limits
plt.xlim(0, 12)
plt.ylim(0, 12)

# Label the plot
plt.title('Graphical Solution to Maximize Cafeteria Meal Revenue')
plt.xlabel('Units of Meal A (x)')
plt.ylabel('Units of Meal B (y)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
