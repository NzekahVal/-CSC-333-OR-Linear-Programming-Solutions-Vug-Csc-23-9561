```python
# importing necessary libraries for proper code execution
import numpy as np
import matplotlib.pyplot as plt
"""
MAXIMIZE PROFIT
            Z=5x+3y
SUBJECT TO

            x+2y≤8
            3x+2y≤12
            x≥0,y≥0 (non-negativity constraints)
"""
# Define the constraints for the bakery problem
def baking_time_constraint(x):
    return (8 - x) / 2  # From x + 2y = 8

def flour_constraint(x):
    return (12 - 3 * x) / 2  # From 3x + 2y = 12

# Generate x values for the plot
x = np.linspace(0, 6, 400)

# Calculate y values for each constraint
y_baking_time = baking_time_constraint(x)
y_flour = flour_constraint(x)

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot the constraint lines
plt.plot(x, y_baking_time, label=r'$x + 2y \leq 8$', color='blue')
plt.plot(x, y_flour, label=r'$3x + 2y \leq 12$', color='green')

# Highlight the feasible region
y_fill = np.minimum(y_baking_time, y_flour)
plt.fill_between(x, y_fill, 0, where=(y_fill >= 0), color='gray', alpha=0.3, label='Feasible Region')

# Set axis limits
plt.xlim(0, 6)
plt.ylim(0, 6)

# Label the plot
plt.title('Graphical Solution to Maximize Bakery Profit')
plt.xlabel('Units of Chocolate Cakes (x)')
plt.ylabel('Units of Vanilla Cakes (y)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
