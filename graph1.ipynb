```python
# importing necessary libraries for proper code execution
import numpy as np
import matplotlib.pyplot as plt


"""
MAXIMIZE
            Z=3x+4y
SUBJECT TO

            2x+3y≤12
            x+2y≤8
            x≥0,y≥0 (non-negativity constraints)
"""

# Define the equations of the constraint lines
def constraint1(x):
    return (12 - 2*x) / 3  # From 2x + 3y = 12

def constraint2(x):
    return (8 - x) / 2  # From x + 2y = 8

# Generate x values for the plot
x = np.linspace(0, 8, 400)

# Calculate y values for each constraint
y1 = constraint1(x)
y2 = constraint2(x)

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot the constraint lines
plt.plot(x, y1, label=r'$2x + 3y \leq 12$', color='blue')
plt.plot(x, y2, label=r'$x + 2y \leq 8$', color='green')

# Highlight the feasible region
y_fill = np.minimum(y1, y2)
plt.fill_between(x, y_fill, 0, where=(y_fill >= 0), color='gray', alpha=0.3, label='Feasible Region')

# Set axis limits
plt.xlim(0, 6)
plt.ylim(0, 5)

# Label the plot
plt.title('Graphical Solution to the Linear Programming Problem')
plt.xlabel('Units of Product A (x)')
plt.ylabel('Units of Product B (y)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
