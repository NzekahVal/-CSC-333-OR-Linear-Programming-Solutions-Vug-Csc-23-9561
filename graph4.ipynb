```python
# importing necessary libraries for proper code execution
import numpy as np
import matplotlib.pyplot as plt

"""
MAXIMIZE
            Z=4x+5y
SUBJECT TO

            x+3y≤20
            x+2y≤15
            x≥0,y≥0 (non-negativity constraints)
"""
# Define the constraints
def budget_constraint(x):
    return (20 - x) / 2  # From x + 2y = 20

def capacity_constraint(x):
    return (15 - x) / 2  # From x + 2y = 15

# Generate x values for the plot
x = np.linspace(0, 20, 400)

# Calculate y values for each constraint
y_budget = budget_constraint(x)
y_capacity = capacity_constraint(x)

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot the constraints
plt.plot(x, y_budget, label=r'$x + 2y \leq 20$', color='blue')
plt.plot(x, y_capacity, label=r'$x + 2y \leq 15$', color='green')

# Highlight the feasible region
y_fill = np.minimum(y_budget, y_capacity)
plt.fill_between(x, y_fill, 0, where=(y_fill >= 0), color='gray', alpha=0.3, label='Feasible Region')

# Set axis limits
plt.xlim(0, 20)
plt.ylim(0, 10)

# Label the plot
plt.title('Graphical Solution to Maximize Revenue')
plt.xlabel('Units of Product A (x)')
plt.ylabel('Units of Product B (y)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
