```python
# importing necessary libraries for proper code execution
import numpy as np
import matplotlib.pyplot as plt

"""
MAXIMIZE PROFIT
            Z=10x+12y
SUBJECT TO

            4x+3y≤30
            x+2y≤18
            3x+2y≤24
            x≥0,y≥0 (non-negativity constraints)
"""
# Define the constraints for the manufacturing problem
def labor_constraint(x):
    return (30 - 4 * x) / 3  # From 4x + 3y = 30

def material_constraint(x):
    return (18 - x) / 2  # From x + 2y = 18

def machine_time_constraint(x):
    return (24 - 3 * x) / 2  # From 3x + 2y = 24

# Generate x values for the plot
x = np.linspace(0, 7, 400)

# Calculate y values for each constraint
y_labor = labor_constraint(x)
y_material = material_constraint(x)
y_machine_time = machine_time_constraint(x)

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot the constraint lines
plt.plot(x, y_labor, label=r'$4x + 3y \leq 30$', color='blue')
plt.plot(x, y_material, label=r'$x + 2y \leq 18$', color='green')
plt.plot(x, y_machine_time, label=r'$3x + 2y \leq 24$', color='red')

# Highlight the feasible region
y_fill = np.minimum(np.minimum(y_labor, y_material), y_machine_time)
plt.fill_between(x, y_fill, 0, where=(y_fill >= 0), color='gray', alpha=0.3, label='Feasible Region')

# Set axis limits
plt.xlim(0, 7)
plt.ylim(0, 10)

# Label the plot
plt.title('Graphical Solution to Maximize Revenue from Manufacturing Products')
plt.xlabel('Units of Product P1 (x)')
plt.ylabel('Units of Product P2 (y)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
